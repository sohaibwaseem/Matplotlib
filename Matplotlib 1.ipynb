```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(np.linspace(0,10,11), np.linspace(0,100,11))
axes.set_xlabel('X-axis')
axes.set_ylabel('y-axis')
axes.set_title('title')
```

```python
x = np.linspace(0,5,11)
y = pow(x, 2)
```

```python
x
```

```python
y
```

```python
plt.subplot(1,2,1) # (number of rows, number of columns, number of plots)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # Big box
axes2 = fig.add_axes([0.3, 0.35, 0.4, 0.4]) # internal box. First two values are the distance from the
                                            # walls of outer box
    
axes1.plot(x,y)
axes1.set_title('Large Figure')
axes2.plot(y,x)
axes2.set_title('Small Figure')
```
