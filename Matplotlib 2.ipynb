```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2) # nrows = number of figs in a row
                                               # ncols = number of figs in a column
    
axes[0].plot(x,y,'r')
axes[0].set_title('First Plot')

axes[1].plot(y,x,'g')
axes[1].set_title('Second Plot')

```

```python
fig.savefig('Desktop/Subplots.png', dpi = 300)
```

```python
axes
```

## Figure size and DPI

```python
fig2 = plt.figure(figsize=(8,2))

ax = fig2.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig1, axes1 = plt.subplots(nrows = 2, ncols = 1, figsize = (8,2))

axes1[0].plot(x,y, 'r')

axes1[1].plot(y,x, 'g')

plt.tight_layout()
```

*Save Graph on you Desktop*

```python
fig1.savefig('Desktop/My_fig.png', dpi = 400)
```

```python
fig3 = plt.figure()

ax1 = fig3.add_axes([0,0,1,1])

ax1.plot(x,x**2, label = "X**2")
ax1.plot(x,x**3, label = "X**3")

plt.title('Map')

plt.legend(loc = 0) # loc values are from 1 to 10
```
